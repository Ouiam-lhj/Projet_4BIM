# Code pour filtrer les visages de profil

```python
import pandas as pd
import numpy as np
import statistics
```

```python
list_landmarks=pd.read_csv("C:/Users/acaci/Downloads/list_landmarks_align_celeba.txt" , sep = "\s+", header = 0)
print(list_landmarks)
```

```python
droping = []
for i in range(len(list_landmarks)):
    mean_eye = statistics.mean(list_landmarks.iloc[i,[0,2]])
    mean_eye_inf = mean_eye - 10
    mean_eye_sup = mean_eye + 10
    mean_mouth = statistics.mean(list_landmarks.iloc[i,[6,8]])
    mean_mouth_inf = mean_mouth - 10
    mean_mouth_sup = mean_mouth + 10
    if list_landmarks.iloc[i,4] < mean_eye_inf or list_landmarks.iloc[i,4] > mean_eye_sup or list_landmarks.iloc[i,4] < mean_mouth_inf or list_landmarks.iloc[i,4] > mean_mouth_sup:
        string_to_app = str(i+1) + ".jpg"
        string_to_app = string_to_app.zfill(10)
        droping.append(string_to_app)
```

```python
print(droping)
len(droping)
list_landmarks_filtered = list_landmarks
list_landmarks_filtered = list_landmarks_filtered.drop(droping)
print(list_landmarks_filtered)
```
