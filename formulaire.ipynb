```python
import customtkinter

def close_window():
    app.destroy()

# Définition des variables
titre = "Questionnaire de pré-sélection"
consignes = "Veuillez répondre aux questions suivantes pour l'identification de l'individu. Si vous avez le moindre doute, cochez la case \"Je ne sais pas\"."
font_size_questions = 16

# Création de la fenêtre
app = customtkinter.CTk()
app.title("Questionnaire")
app.geometry("520x600")
app.grid_columnconfigure(0, weight=1)
app.grid_rowconfigure(0, weight=1)

# Scrollable frame
scroll = customtkinter.CTkScrollableFrame(app, width=500, height=600)
scroll.grid(row=0, column=0, columnspan=3, sticky="nsew")

# Configuration de la grille
scroll.grid_columnconfigure(0, weight=1)

# Création des widgets
# Titres
label_titre = customtkinter.CTkLabel(scroll, text=titre, font=("Arial", 30), wraplength=480)
label_titre.grid(row=0, column=0, columnspan=3, pady=(20,10))

# Consignes avec wraplength
label_consignes = customtkinter.CTkLabel(scroll, text=consignes, font=("Arial", 18), wraplength=480)
label_consignes.grid(row=1, column=0, columnspan=3, pady=(10, 20))

# Ligne de séparation
separator = customtkinter.CTkFrame(scroll, height=2, fg_color="gray")
separator.grid(row=2, column=0, columnspan=3, sticky="ew", pady=(0, 20))

# Questions
text_sex = customtkinter.CTkLabel(scroll, text="Quel était le sexe de l'individu ?", font=("Arial", font_size_questions))
text_sex.grid(row=3, column=0, padx=20, pady=(20, 2))

text_peau = customtkinter.CTkLabel(scroll, text="Quelle était la couleur de peau de l'individu ?", font=("Arial", font_size_questions))
text_peau.grid(row=5, column=0, padx=20, pady=(20, 2))

text_cheveux = customtkinter.CTkLabel(scroll, text="Quelle était la couleur des cheveux de l'individu ?", font=("Arial", font_size_questions))
text_cheveux.grid(row=7, column=0, padx=20, pady=(20, 2))

text_texture = customtkinter.CTkLabel(scroll, text="Quel était la texture de cheveux de l'individu ?", font=("Arial", font_size_questions))
text_texture.grid(row=9, column=0, padx=20, pady=(20, 2))

text_corpu = customtkinter.CTkLabel(scroll, text="Quelle était la corpulence de l'individu ?", font=("Arial", font_size_questions))
text_corpu.grid(row=11, column=0, padx=20, pady=(20, 2))

text_lunettes = customtkinter.CTkLabel(scroll, text="L'individu portait-il des lunettes ?", font=("Arial", font_size_questions))
text_lunettes.grid(row=13, column=0, padx=20, pady=(20, 2))

text_age = customtkinter.CTkLabel(scroll, text="L'individu vous paraissait plutôt :", font=("Arial", font_size_questions))
text_age.grid(row=15, column=0, padx=20, pady=(20, 2))

text_pilosité = customtkinter.CTkLabel(scroll, text="Quelle était la pilosité de l'individu", font=("Arial", font_size_questions))
text_pilosité.grid(row=17, column=0, padx=20, pady=(20, 2))

# Réponses
combobox_sex = customtkinter.CTkComboBox(scroll, values=["Homme", "Femme", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_sex.grid(row=4, column=0, padx=20, pady=(2, 20))
combobox_sex.set("Sélectionner")

combobox_peau = customtkinter.CTkComboBox(scroll, values=["Pâle", "Foncée", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_peau.grid(row=6, column=0, padx=20, pady=(2, 20))
combobox_peau.set("Sélectionner")

combobox_cheveux = customtkinter.CTkComboBox(scroll, values=["Noirs/Bruns/Châtains", "Blonds", "Roux", "Chauve", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_cheveux.grid(row=8, column=0, padx=20, pady=(2, 20))
combobox_cheveux.set("Sélectionner")

combobox_texture = customtkinter.CTkComboBox(scroll, values=["Lisses", "Bouclés", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_texture.grid(row=10, column=0, padx=20, pady=(2, 20))
combobox_texture.set("Sélectionner")

combobox_corpu = customtkinter.CTkComboBox(scroll, values=["Faible", "Forte", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_corpu.grid(row=12, column=0, padx=20, pady=(2, 20))
combobox_corpu.set("Sélectionner")

combobox_lunettes = customtkinter.CTkComboBox(scroll, values=["Oui", "Non", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_lunettes.grid(row=14, column=0, padx=20, pady=(2, 20))
combobox_lunettes.set("Sélectionner")

combobox_age = customtkinter.CTkComboBox(scroll, values=["Jeune", "Agé", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_age.grid(row=16, column=0, padx=20, pady=(2, 20))
combobox_age.set("Sélectionner")

combobox_pilosité = customtkinter.CTkComboBox(scroll, values=["Absente", "Bouc", "Moustache", "Bouc et moustache", "Je ne sais pas"], width=200, height=20, font=("Arial", font_size_questions))
combobox_pilosité.grid(row=18, column=0, padx=20, pady=(2, 20))
combobox_pilosité.set("Sélectionner")

# Bouton de validation
button = customtkinter.CTkButton(scroll, text="Valider", font=("Arial", font_size_questions), width=200, height=20,command=close_window)
button.grid(row=19, column=0, pady=50)

app.mainloop()
```
