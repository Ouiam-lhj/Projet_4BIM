```python
import customtkinter
from tkinter import messagebox

def close_window():
    # Récupération des réponses
    reponses = {
        "Sexe": combobox_sex.get(),
        "Couleur de peau": combobox_peau.get(),
        "Couleur des cheveux": combobox_cheveux.get(),
        "Texture des cheveux": combobox_texture.get(),
        "Corpulence": combobox_corpu.get(),
        "Lunettes": combobox_lunettes.get(),
        "Âge": combobox_age.get(),
        "Pilosité": combobox_pilosité.get()
    }
    if "Sélectionner" in reponses.values():
        messagebox.showerror("", "Veuillez répondre à toutes les questions.")
    else:
        messagebox.showinfo("", "Vos réponses ont bien été enregistrées.")
        app.destroy()

# Définition des variables
titre = "Questionnaire de pré-sélection"
consignes = "Veuillez répondre aux questions suivantes pour l'identification de l'individu. Si vous avez le moindre doute, cochez la case \"Je ne sais pas\".\n\nN'oubliez pas de sauvegarder vos réponses en cliquant sur le bouton \"Valider\" en bas de page."
font_size_questions = 16

# Création de la fenêtre
customtkinter.set_appearance_mode("light")  # Mode clair
customtkinter.set_default_color_theme("blue")  # Thème bleu par défaut

app = customtkinter.CTk()
app.title("Questionnaire")
app.geometry("520x600")
app.configure(bg="#f5f5f5")  # Couleur de fond de la fenêtre
app.grid_columnconfigure(0, weight=1)
app.grid_rowconfigure(0, weight=1)

# Scrollable frame
scroll = customtkinter.CTkScrollableFrame(app, width=500, height=600, fg_color="#ffffff")  # Fond blanc
scroll.grid(row=0, column=0, columnspan=3, sticky="nsew")

# Configuration de la grille
scroll.grid_columnconfigure(0, weight=1)

# Création des widgets
# Titres
label_titre = customtkinter.CTkLabel(scroll, text=titre, font=("Arial", 30), wraplength=480, text_color="#333333")  # Texte gris foncé
label_consignes = customtkinter.CTkLabel(scroll, text=consignes, font=("Arial", 18), wraplength=480, text_color="#555555")  # Texte gris moyen
separator = customtkinter.CTkFrame(scroll, height=2, fg_color="#cccccc")  # Séparateur gris clair

# Questions
text_color = "#333333"  # Couleur du texte des questions
text_sex = customtkinter.CTkLabel(scroll, text="Quel était le sexe de l'individu ?", font=("Arial", font_size_questions), text_color=text_color)
text_peau = customtkinter.CTkLabel(scroll, text="Quelle était la couleur de peau de l'individu ?", font=("Arial", font_size_questions), text_color=text_color)
text_cheveux = customtkinter.CTkLabel(scroll, text="Quelle était la couleur des cheveux de l'individu ?", font=("Arial", font_size_questions), text_color=text_color)
text_texture = customtkinter.CTkLabel(scroll, text="Quel était la texture de cheveux de l'individu ?", font=("Arial", font_size_questions), text_color=text_color)
text_corpu = customtkinter.CTkLabel(scroll, text="Quelle était la corpulence de l'individu ?", font=("Arial", font_size_questions), text_color=text_color)
text_lunettes = customtkinter.CTkLabel(scroll, text="L'individu portait-il des lunettes ?", font=("Arial", font_size_questions), text_color=text_color)
text_age = customtkinter.CTkLabel(scroll, text="L'individu vous paraissait plutôt :", font=("Arial", font_size_questions), text_color=text_color)
text_pilosité = customtkinter.CTkLabel(scroll, text="Quelle était la pilosité de l'individu", font=("Arial", font_size_questions), text_color=text_color)

# Réponses (Création et set)
combobox_style = {"width": 200, "height": 20, "font": ("Arial", font_size_questions), "state": "readonly"}
combobox_sex = customtkinter.CTkComboBox(scroll, values=["Homme", "Femme", "Je ne sais pas"], **combobox_style)
combobox_sex.set("Sélectionner")

combobox_peau = customtkinter.CTkComboBox(scroll, values=["Pâle", "Foncée", "Je ne sais pas"], **combobox_style)
combobox_peau.set("Sélectionner")

combobox_cheveux = customtkinter.CTkComboBox(scroll, values=["Noirs/Bruns/Châtains", "Blonds", "Roux", "Chauve", "Je ne sais pas"], **combobox_style)
combobox_cheveux.set("Sélectionner")

combobox_texture = customtkinter.CTkComboBox(scroll, values=["Lisses", "Bouclés", "Je ne sais pas"], **combobox_style)
combobox_texture.set("Sélectionner")

combobox_corpu = customtkinter.CTkComboBox(scroll, values=["Faible", "Forte", "Je ne sais pas"], **combobox_style)
combobox_corpu.set("Sélectionner")

combobox_lunettes = customtkinter.CTkComboBox(scroll, values=["Oui", "Non", "Je ne sais pas"], **combobox_style)
combobox_lunettes.set("Sélectionner")

combobox_age = customtkinter.CTkComboBox(scroll, values=["Jeune", "Agé", "Je ne sais pas"], **combobox_style)
combobox_age.set("Sélectionner")

combobox_pilosité = customtkinter.CTkComboBox(scroll, values=["Absente", "Bouc", "Moustache", "Bouc et moustache", "Je ne sais pas"], **combobox_style)
combobox_pilosité.set("Sélectionner")

# Bouton de validation
button = customtkinter.CTkButton(scroll, text="Valider", font=("Arial", font_size_questions), width=200, height=30, command=close_window, fg_color="#528868", text_color="#ffffff")  # Bouton bleu avec texte blanc

# Placement des widgets
label_titre.grid(row=0, column=0, columnspan=3, pady=(20, 10))
label_consignes.grid(row=1, column=0, columnspan=3, pady=(10, 20))
separator.grid(row=2, column=0, columnspan=3, sticky="ew", pady=(0, 20))

text_sex.grid(row=3, column=0, padx=20, pady=(20, 2))
combobox_sex.grid(row=4, column=0, padx=20, pady=(2, 20))

text_peau.grid(row=5, column=0, padx=20, pady=(20, 2))
combobox_peau.grid(row=6, column=0, padx=20, pady=(2, 20))

text_cheveux.grid(row=7, column=0, padx=20, pady=(20, 2))
combobox_cheveux.grid(row=8, column=0, padx=20, pady=(2, 20))

text_texture.grid(row=9, column=0, padx=20, pady=(20, 2))
combobox_texture.grid(row=10, column=0, padx=20, pady=(2, 20))

text_corpu.grid(row=11, column=0, padx=20, pady=(20, 2))
combobox_corpu.grid(row=12, column=0, padx=20, pady=(2, 20))

text_lunettes.grid(row=13, column=0, padx=20, pady=(20, 2))
combobox_lunettes.grid(row=14, column=0, padx=20, pady=(2, 20))

text_age.grid(row=15, column=0, padx=20, pady=(20, 2))
combobox_age.grid(row=16, column=0, padx=20, pady=(2, 20))

text_pilosité.grid(row=17, column=0, padx=20, pady=(20, 2))
combobox_pilosité.grid(row=18, column=0, padx=20, pady=(2, 20))

button.grid(row=19, column=0, pady=(50, 20))

app.mainloop()
```
